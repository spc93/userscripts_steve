# convert focus size to slope errors
## assume linear relationship, e.g. no other broadening of focus (good for large errors)

Dt = slope error

Df = focus size

virtual source size V = 2 * Dt * M

demagnification by mirrors = (F-m1)/m1 (close to 1)

Df = V * (F-m1)/m1

Df = 2 * Dt * M * (F-m1)/m1

Dt = Df / M /(F/m1 - 1)/2
