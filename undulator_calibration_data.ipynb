```python
import numpy as np
from matplotlib.pyplot import *
```

```python
file = '/home/spc93/undulator_i16/data_7_6_2019/ucalibrate.log'
```

```python
with open(file) as f:
    A = np.loadtxt((x.replace('-',' ') for x in f)).transpose()

date = []
for i in range(A.shape[1]):
    date += [dt.datetime(int(A[2][i])+2000, int(A[1][i]), int(A[0][i]))]
    
harmonic, energy, gap, offset = A[3], A[4], A[5], A[6]
```

```python
figure()
plot(date, offset,'.'); ylim(0.4,0.6); ylabel('Offset')
```

```python
figure()
for harm in range(3,13):
    select = harmonic==harm
    plot(energy[select], gap[select], '.')
```
