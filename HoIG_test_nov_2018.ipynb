```python
sx=[]; sy=[]; sumcts=[]; wid=[]
for scan in range(724443, 724521):
    n = pdnx(p % scan)
    #plot(n.eta, n['sum'])
    try:
        pv_c.fit(n.eta, n['sum'])
        sumcts+=[pv_c.p[0]]
        wid+=[pv_c.p[2]]
    except: # in case no fit
        sumcts+=[0]
        wid+=[0]    
        
    sx+=[float(n.nx.entry1.before_scan.positions.sx)]
    sy+=[float(n.nx.entry1.before_scan.positions.sy)]
    
msx, msy, msum = vec2mat(sx, sy, sumcts)
figure(); gca(projection='3d').plot_surface(msx, msy, msum)   
title('sum_counts Sample A eta scans vs sx, sy'); xlabel('sx'); ylabel('sy'); 

msx, msy, mwid = vec2mat(sx, sy, wid)
figure(); gca(projection='3d').plot_surface(msx, msy, mwid)   
title('fwhm Sample A eta scans vs sx, sy'); xlabel('sx'); ylabel('sy');

#for i in range(len(sx)):
#    print '%.2f\t%.2f\t%.0f\t%.4f' % (sx[i], sy[i], sumcts[i], wid[i])
```
