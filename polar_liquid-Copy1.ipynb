```python
import numpy as np
%matplotlib nbagg
from matplotlib.pyplot import *
```

```python
atoms=[
        {'name':'O1', 'z_eff':8.0, 'posn':0.5},
        {'name':'O2', 'z_eff':8.0, 'posn':1.0},
        {'name':'O3', 'z_eff':8.0, 'posn':1.5},
        {'name':'Br1', 'z_eff':35.0+10.0J, 'posn':0.0},
      ]

BromoButane_vert=[
        {'name':'c1', 'z_eff':6.0, 'posn':-2.4758},
        {'name':'c2', 'z_eff':6.0, 'posn':-1.96},
        {'name':'c3', 'z_eff':6.0, 'posn':-2.4784},
        {'name':'c4', 'z_eff':6.0, 'posn':-1.4014},
        {'name':'Br1', 'z_eff':35.0+10.0J, 'posn':0.0},
      ]

BromoButane_horiz=[
        {'name':'c1', 'z_eff':6.0, 'posn':-1.4462},
        {'name':'c2', 'z_eff':6.0, 'posn':0},
        {'name':'c3', 'z_eff':6.0, 'posn':1.4427},
        {'name':'c4', 'z_eff':6.0, 'posn':2.5356},
        {'name':'Br1', 'z_eff':35.0+10.0J, 'posn':0.0},
      ]
```

```python
def blah(dict):
    tth_deg=np.arange(0,60.0,.1)
    lam=12.4/33.3
    #q=np.arange(.1,5.0,.05)
    q=2*np.pi*np.sin(tth_deg/2*np.pi/180)/lam
    fp=0; fn=0;
    for atom in dict:
       fp+=atom['z_eff']*np.exp(1.0J*q*atom['posn'])
       fn+=atom['z_eff']*np.exp(-1.0J*q*atom['posn'])
    Ip=np.abs(fp)**2
    In=np.abs(fn)**2
    #print Ip, In
    figure(); plot(tth_deg,Ip,'r'); plot(tth_deg,In,'g'); xlabel('tth (deg)'); ylabel('Intensity (form factor**2)'); grid(1)

blah(atoms)
blah(BromoButane_vert)
blah(BromoButane_horiz)
```
